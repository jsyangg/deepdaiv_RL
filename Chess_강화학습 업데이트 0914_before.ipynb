```python
# 개별 폰 움직임 (6개)

def black_move_down(i,j) :
    if j==3 :
        return "Unavailable Movement"
    elif board[i][j][0] == "-"  :
        return "Unavailable Movement"
    elif board[i][j][0] == "W"  :
        return "Unavailable Movement"    
    elif board[i][j+1][0] == "W"  :
        return "Unavailable Movement"   
    elif board[i][j+1][0] == "B"  :
        return "Unavailable Movement"     
    elif board[i][j][0] == "B" and board[i][j+1][0] == "-":
        board[i][j+1]=board[i][j]
        board[i][j] = "--"
        return "black_move_down"
    
def white_move_up(i,j):
    if j==0 :
        return "Unavailable Movement"
    elif board[i][j][0] == "-"  :
        return "Unavailable Movement"
    elif board[i][j][0] == "B"  :
        return "Unavailable Movement"    
    elif board[i][j-1][0] == "W"  :
        return "Unavailable Movement"
    elif board[i][j-1][0] == "B"  :
        return "Unavailable Movement"  
    elif board[i][j][0] == "W" and board[i][j-1][0] == "-":
        board[i][j-1]=board[i][j]
        board[i][j] = "--"
        return "white_move_up"
    
def black_attack_left(i,j):
    if i == 0 :
        return "Unavailable Movement"
    elif j == 3 :
        return "Unavailable Movement"
    elif board[i][j][0] == "-"  :
        return "Unavailable Movement"
    elif board[i][j][0] == "W"  :
        return "Unavailable Movement"   
    elif board[i-1][j+1][0] == "W":
        board[i-1][j+1]=board[i][j]
        board[i][j]="--"
        return "black_attack_left"
    elif board[i-1][j+1][0] == "-":
        return "Unavailable Movement"
    elif board[i-1][j+1][0] == "B":
        return "Unavailable Movement"
    
def black_attack_right(i,j):
    if i == 3 :
        return "Unavailable Movement"
    elif j == 3 :
        return "Unavailable Movement"
    elif board[i][j][0] == "-"  :
        return "Unavailable Movement"
    elif board[i][j][0] == "W"  :
        return "Unavailable Movement"   
    elif board[i+1][j+1][0] == "W":
        board[i+1][j+1]=board[i][j]
        board[i][j]="--"
        return "black_attack_right"
    elif board[i+1][j+1][0] == "-":
        return "Unavailable Movement"
    elif board[i+1][j+1][0] == "B":
        return "Unavailable Movement"
    
def white_attack_left(i,j):
    if i == 0 :
        return "Unavailable Movement"
    elif j == 0 :
        return "Unavailable Movement"
    elif board[i][j][0] == "-" :
        return "Unavailable Movement"
    elif board[i][j][0] == "B"  :
        return "Unavailable Movement"   
    elif board[i-1][j-1][0] == "B":
        board[i-1][j-1]=board[i][j]
        board[i][j]="--"
        return "white_attack_left"
    elif board[i-1][j-1][0] == "-":
        return "Unavailable Movement"
    elif board[i-1][j-1][0] == "W":
        return "Unavailable Movement"
    
def white_attack_right(i,j):
    if i == 3 :
        return "Unavailable Movement"
    elif j == 0 :
        return "Unavailable Movement"
    elif board[i][j][0] == "-" :
        return "Unavailable Movement"
    elif board[i][j][0] == "B"  :
        return "Unavailable Movement"    
    elif board[i+1][j-1][0] == "B":
        board[i+1][j-1]=board[i][j]
        board[i][j]="--"
        return "white_attack_right"
    elif board[i+1][j-1][0] == "-":
        return "Unavailable Movement"
    elif board[i+1][j-1][0] == "W":
        return "Unavailable Movement"
```
